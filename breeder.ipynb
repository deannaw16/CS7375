```python
import pandas as pd
data_path = 'https://raw.githubusercontent.com/deannaw16/CS7375/refs/heads/main/AI%20Project%20Dataset%20Complete.csv'
df = pd.read_csv(data_path, header=0)

columns_to_transform = ['Colonial Breeder', 'Non-Migrant', 'Locally Migrant', 'Long Distance Migrant']
df[columns_to_transform] = df[columns_to_transform].map(lambda x: 1 if x is True else (0 if x is False else str(x)))

df.head()
```

```python
df_selected = df[columns_to_transform]
df_selected.head()
```

```python
df_selected.to_csv('breeder_true_or_false.csv', index=False)
```
